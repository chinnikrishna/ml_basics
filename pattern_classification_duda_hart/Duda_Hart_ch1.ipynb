# Introduction

To illustrate the problem of Pattern Classification, they consider an example of classifying fishes (Salmon vs SeaBass)

1. To get to features of the fish, the image from camera is segmented to separate the fish from background and other fishes (extraction). From this we measure characteristic property of the fish (say length). These properties are called features and are input to the classification system.

2. Let us suppose that we can classify the fish based on one feature alone (length in this case). There is a <strong>cost</strong> associated with mis-classifying the fish which is not equal in most of the cases.

    For example if we classify salmon(which is expensive) as sea-bass would not bother the customer, so this mis-classification is ok. But classifying sea-bass as salmon would have a negative effect on brand of the company (They are replacing salmon with cheaper fish). 
    
    So given a feature, how to classify falls under the purview of decision theory - it helps to define the right decision boundary to separate the classes.
    
3. Another aspect to consider is how well this classification system works on a new fish images, called "generalization". <strong> Pattern Classification at its heart is the task of recovering the model that generated the patterns</strong>. This is the central problem addressed by statistical learning theory. 

## Importance of Representation
A central aspect in virtually every pattern recognition problem is that of achieving such a good representation, 

1) one in which the structural relationships among the components is simply and naturally revealed

2) one in which the true (unknown) model ofthe patterns can be expressed. In some cases patterns should be represented as vectors ofreal-valued numbers, in others ordered lists ofattributes, in yet others descriptions ofparts and their relations, and so forth. 

3) We seek a representation in which the patterns that lead to the same action are somehow close to one another, yet far from those that demand a different action. 

The extent to which we create or learn a proper representation and how we quantify near and far apart will determine the success ofour pattern classifier.

Consider the example of recognizing a chair. Chairs can come in large number of varieties but they all can be clustered together based on their functionality.
